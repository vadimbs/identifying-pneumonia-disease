## Modeling

#### Variables

```python
data_dir = 'chest_xray/'
data_dir_rsmpld = 'chest_xray_resampled/'
batch_size = 20

model_metrics = {
    'Accuracy': 'acc',
    #'TruePositives': metrics.TruePositives(),
    #'TrueNegatives': metrics.TrueNegatives(),
    #'FalsePositives': metrics.FalsePositives(),
    'FalseNegatives': metrics.FalseNegatives(),
    #'Precision': metrics.Precision(),
    'Recall': metrics.Recall(),
    #'AUC': metrics.AUC(),
}
```

### Build a Baseline CNN

```python
img_size = (150, 150)
epochs = 20
max_steps = 50

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(img_size[0], img_size[1], 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=list(model_metrics.values())
             )

model, history = process_model('base',
                               model, 
                               model_metrics, 
                               data_dir_rsmpld, 
                               img_size=img_size, 
                               batch_size=batch_size, 
                               epochs=epochs,
                               max_steps=max_steps)
```

Training (20 epochs) took 8 mins

### Loading a Pretrained CNN

```python
def pretrained_model_VGG19(model_metrics, 
                           img_size = (150, 150), 
                           epochs=10, 
                           max_steps=50, 
                           trainable_layer=None, 
                           lr=1e-4, 
                           aug=False):
    
    model_name = 'VGG19_' + str(trainable_layer or '') + str('_AUG' if aug is not False else '')
    
    # Load pretrained model
    cnn_base = VGG19(weights='imagenet',
                     include_top=False,
                     input_shape=(img_size[0], img_size[1], 3))


    # Define model architecture
    model = models.Sequential()
    model.add(cnn_base)
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))

    if trainable_layer is None:
        cnn_base.trainable = False
        
    else:
        cnn_base.trainable = True
        set_trainable = False
        for layer in cnn_base.layers:
            if layer.name == trainable_layer:
                set_trainable = True
            if set_trainable:
                layer.trainable = True
            else:
                layer.trainable = False


    model.compile(loss='binary_crossentropy',
                  optimizer=optimizers.RMSprop(lr=lr),
                  metrics=list(model_metrics.values())
                 )

    model, history = process_model(model_name,
                                   model, 
                                   model_metrics,
                                   data_dir_rsmpld, 
                                   img_size=img_size, 
                                   batch_size=batch_size, 
                                   epochs=epochs,
                                   max_steps=max_steps,
                                   aug=aug)
```

```python
pretrained_model_VGG19(model_metrics)
```

Training (10 epochs) took 21 minutes<br>
Comparing to baseline model accuracy increased from almost 75% to 80%<br>
Recall remained at the same level at 99%<br>
Sensetivity slightly increased from 95% to 96%

Let's increase image size from 150x150 to 240x240
